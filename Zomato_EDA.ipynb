Here in the data,Cusines is the feature that contains 9 missing values

```python
#Heat map to see the missing values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

Merging another excel file as the dataset contains same feature in the dataset i.e Country_Code

```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_country.describe()
```

Mering df and df_country data sets

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

To check the datatype of this

```python
final_df.dtypes
```

pie chart

```python
country_val = final_df.Country.value_counts().values
```

```python
country_names=final_df.Country.value_counts().index
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato has the maximum records or transactions from India,after that Unites States and then United Kindom.
